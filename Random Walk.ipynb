```python
import numpy as np
import matplotlib.pyplot as plt
```

### Discrete Stochastic Process: Random Walk

```python
# Basic Setup
T = 1 # End Time
t = np.linspace(0, T, 100) # each time point within the time interval
dt = t[1] - t[0] # time grid
Nsims = 100 # number of simulations
```

The discrete-time stochastic process random walk is defined as:

$$X_k = X_{k-1} + \sqrt{\Delta t}*z_{k}$$

where $\sqrt{\Delta t}*z_{k}$ is the Winer process $dW_{t}$, and z$\sim_{\mathbb{P}} N(0, 1)$

```python
# Random Walk
def RW(Nsims, t):
    
    w = np.zeros((Nsims, len(t)))
    
    for i in range(len(t)-1):
        w[:, i+1] = w[:, i] + np.sqrt(dt)*np.random.randn(Nsims)
    
    return w
```

```python
# Stock Price
S = RW(Nsims, t)
```

```python
# plot 10 price simulations
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.plot(t, S[i,:])
plt.xlabel(r"Time t")
plt.ylabel(r"Stock price S")
plt.show()
```
