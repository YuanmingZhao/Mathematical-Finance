```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# discrete time
```

```python
T = 1
t = np.linspace(0, T, 100)
dt = t[1] - t[0]
Nsims = 100
```

```python
def RW(Nsims, t):
    
    w = np.zeros((Nsims, len(t)))
    
    for i in range(len(t)-1):
        w[:, i+1] = w[:, i] + np.sqrt(dt)*np.random.randn(Nsims)
    
    return w
```

```python
S = RW(Nsims, t)
```

```python
for i in range(10):
    plt.plot(t, S[i,:])
plt.show()
```

```python
# continously stochastic process
## Geometric Brownian Motion
```

```python
# basic setup
mu = 0.1
sigma = -0.2
S0 = 100
```

```python
def GBM_(Nsims, t):
    S = np.zeros((Nsims, len(t)))
    S[:, 0] = S0
    
    for i in range(len(t) - 1):
        S[:, i+1] = S[:, i]*np.exp((mu - 0.5*sigma**2)*dt + sigma*np.random.randn(Nsims)*np.sqrt(dt))
        
    return S
```

```python
S_ = GBM_(Nsims, t)
```

```python
for i in range(10):
    plt.plot(t, S_[i, :])
plt.show()
```
